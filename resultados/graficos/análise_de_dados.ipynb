```python
#4.2) ESCOLHER UM SITE ESPECÍFICO DA BASE PARA VER O RISCO
plt.rcParams.update({'font.size': 12})
plt.rcParams['figure.figsize'] = (6.4, 6.8)
risco_site = 0
for d in dados['Sites']:
    if d == 'acontecaeventos':
        dados_site = dados['Risco']
        dados_site = dados_site[risco_site]
        print(dados_site)
    risco_site+=1

array_enfeite = [1]
risco = []
for i in array_enfeite:
    risco.append(dados_site)

    niveis = risco
    altos = []
    medios = []
    baixos = []
    for i in niveis:
        if i == 3:
            altos.append(i)
        elif i == 2:
            medios.append(i)
        elif i == 1:
            baixos.append(i)

    resultado_dados = [len(altos), len(medios), len(baixos)]

print('altos:  '+ str(len(altos)))
print('médios: '+ str(len(medios)))
print('baixos: '+ str(len(baixos)))

altos = resultado_dados[0]
medios = resultado_dados[1]
baixos = resultado_dados[2]

labels = ['Risco']
x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, altos, width,label = 'altos', color= 'red')
rects2 = ax.bar(x + width/2, medios, width,label = 'medios', color = 'yellow')
rects3 = ax.bar(x + width/4, baixos, width,label = 'baixos', color = 'green')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel("Quantidade de sites")
ax.set_title('Risco dos site \n')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

fig.tight_layout()

plt.show()        
```
